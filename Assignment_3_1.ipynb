```python
import numpy as np
import pandas as pd
import sys
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
def build_window(x,y,size,mat):
  half = size//2
  window = np.zeros([size,size], dtype=int)
  i=0
  for r in range(x-half,x+half+1):
    j=0
    for c in range(y-half,y+half+1):
      if r<0 or c<0 or r>=mat.shape[0] or c>=mat.shape[1]:
        j=j+1
        continue
      else:
        window[i][j] = mat[r][c]
        j=j+1
    i=i+1
  return window
```

```python
def match_method(mat1, mat2, method):
  sum = 0
  if method == "SAD":
    for i in range(mat1.shape[0]):
      for j in range(mat2.shape[1]):
        sum = sum + abs(mat1[i][j]-mat2[i][j])
  else:
    for i in range(mat1.shape[0]):
      for j in range(mat2.shape[1]):
        sum = sum + pow(mat1[i][j]-mat2[i][j],2)
  return sum
  # if method=="SAD":
  #   return np.sum(abs(np.subtract(mat1,mat2)))
  # else:
  #   return np.sum(np.square(np.subtract(mat1,mat2)))
```

```python
def blockMatching(img,img2, window_size, method, offset):
  disparity = np.zeros((img.shape[0],img.shape[1]), dtype=int)
  for i in range(img.shape[0]):
    for j in range(img.shape[1]):
      main_window = build_window(i,j,window_size,img)
      min_match = sys.maxsize
      min_ind = -1
      start = 0
      if j-offset > 0:
        start = j-offset
      end = j+offset
      if end > img2.shape[1]:
        end = img2.shape[1]
      for k in range(start,end):
        match_window = build_window(i,k,window_size,img2)
        matched = match_method(main_window,match_window,method)
        if matched < min_match:
          min_match = matched
          min_ind = k
      disparity[i][j] = abs(min_ind-j)
  return disparity
```

## Stereo Matching 

```python
img1 = cv2.imread("l1.png",cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("r1.png",cv2.IMREAD_GRAYSCALE)
```

###Dynamic Programming

```python
def stereo_matching_iter(img1, img2, c0 = 1, sigma = 2):
  rows = img1.shape[0]
  cols = img1.shape[1]
  img=np.zeros((rows,cols))
  for i in range(rows):
    #mapping for backtracking
    map = np.zeros((cols, cols))
    d = np.zeros((cols, cols))
    #initializing
    for j in range(cols):
      d[j][0] = j*c0
      d[0][j] = j*c0
    #for each row find the min cost (the highest match between cols)
    for j in range(cols):
      for k in range(cols):  
        djk=((img1[i][j]-img2[i][k])**2)/(sigma**2)
        m1 = d[j-1][k-1] + djk
        m2 = d[j-1][k] + c0
        m3 = d[j][k-1] + c0
        d[j][k] = min(m1, m2, m3)
        if(m1 == d[j][k]):
          map[j][k] = 0
        elif(m2 == d[j][k]):
          map[j][k] = 1
        elif(m3 == d[j][k]):
          map[j][k] = 2
    #backtracking
    l = r = cols - 1
    while (l != 0 and r != 0):
      if(map[l][r] == 0):
          #store the desparity
          img[i][l] = np.absolute(l-r)
          l -= 1
          r -= 1
      elif(l != 0 and map[l][r] == 1):
          # img[i][l] = 0
          l -=1
      elif(l != 0 and map[l][r] == 2):
          #right is skipped 
          r -= 1
  return img 
```

```python
res3 = stereo_matching_iter(img1,img2)
cv2_imshow(res3)
```
